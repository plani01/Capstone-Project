```python
"""Suggestion for testing: don't have to resize necesarily or even decrease more, keep base resolution, prob more accurate"""
def Create_Data_Matrix(directory):
    i=0
    for file in os.listdir(directory):
        filename = directory+'/'+file

        # Loads image, converts to grayscale and resizes it to a 300x300 image
        y = np.array(Image.open(filename).convert('RGB').resize((300,300)))

        # Resizes 300x300 image to 90,000x1 array
        col_y = y.ravel()[:,np.newaxis]

        # Saves
        if i==0:
            data = col_y
        else:
            data = np.hstack((data, col_y))

        # Plots image
        """
        plt.figure(figsize=(5,5))
        plt.imshow(y, cmap='gray')
        plt.xticks([]),plt.yticks([])
        plt.show();
        """
        i+=1
        
    return data
```

```python
def Create_Labels_Matrix(directory):
    i=0
    for file in os.listdir(directory):
        filename = directory+'/'+file
        tree = ET.parse(filename)
        root = tree.getroot()
        sample_annotations = []
        for neighbor in root.iter('bndbox'):
            xmin = int(neighbor.find('xmin').text)
            ymin = int(neighbor.find('ymin').text)
            xmax = int(neighbor.find('xmax').text)
            ymax = int(neighbor.find('ymax').text)
            sample_annotations.append([xmin, ymin, xmax, ymax])
            if i==0:
                data = sample_annotations
            else:
                data = np.vstack((data, sample_annotations))
            break;
        i+=1
        
    return data
```

```python
"""X : FEATURE MATRIX For CLASSIFICATION""" 
# transpose because scikit expects matrix as: (samples x eachSample'sData)
X = np.hstack((bottleCap_array, lighter_array, emptyClass_array)).T 

# stack all arrays together, no requirment to have same amount of images per class. But of course more images
# necessary for 'empty class', only 67 now 
X.shape
```

```python
"""Train Test Split for Classfication"""
# Normalization: divide by pixel max value then all values [0,1]
X =X/255.

# here I need to setup Cross Validation, look back at notes, vary the kernel function for the model below 
X_train, X_test, t_train, t_test = train_test_split(X, t, test_size=0.2, random_state=42, stratify=t)

X_train.shape, X_test.shape
```

```python
"""MODEL TRAINING For Classifier """

# Classificiatoin model is created here. I will need a seperate SVM(SVC) model for regression(bbox prediction)
# She also gave possible idea for future: split input image incoming from robot's camera. Only look in the immediate forground.
# In this foreground, detect if trash or not is present: assumes objects will only be feed sequentially(no two trash objects
#   will sit next to each other, then a decison would need to be made to pick between them). In this case, no regression or 
#  bounding box will be needed at all. 

# I took extra notes about model paramaters(C and epsilon), found in my notebook
model_class = SVC(kernel='rbf', C=1E9) #hard-margin will overfit (always)
model_class.fit(X_train, t_train)
```
